```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
rain_race = pd.read_csv('../data/rain_race.csv')
lando_laps = pd.read_csv('../data/lando_laps.csv')
```

```python
rain_list = rain_race.session_key.to_list()
```

```python
lando_rain = lando_laps[lando_laps.session_key.isin(rain_list)]
```

```python
lando_rain = lando_rain[['session_key','driver_number','i1_speed','i2_speed','st_speed','lap_duration','duration_sector_1','duration_sector_2','duration_sector_3']]
```

```python
lando_rain = lando_rain.reset_index(drop=True)
lando_rain
```

```python
lando_rain_avg = lando_rain.groupby(['session_key']).agg({'i1_speed':'mean','i2_speed':'mean','st_speed':'mean','lap_duration':'mean','duration_sector_1':'mean','duration_sector_2':'mean','duration_sector_3':'mean'})
```

```python
lando_rain_avg = lando_rain_avg.reset_index()
lando_rain_avg
```

```python
rain_race = rain_race[['location','session_key','year','air_temperature','humidity','pressure','rainfall','track_temperature','wind_speed']]
rain_race
```

```python
lando_rain_avg = lando_rain_avg.merge(rain_race, on = 'session_key', how = 'inner')
```

```python
albon_laps = pd.read_csv('../data/albon_laps.csv')
albon_laps = albon_laps[albon_laps.session_key.isin(rain_list)]
albon_rain = albon_laps[['session_key','driver_number','i1_speed','i2_speed','st_speed','lap_duration','duration_sector_1','duration_sector_2','duration_sector_3']]
albon_rain = albon_rain.reset_index(drop=True)
albon_rain_avg = albon_rain.groupby(['session_key']).agg({'i1_speed':'mean','i2_speed':'mean','st_speed':'mean','lap_duration':'mean','duration_sector_1':'mean','duration_sector_2':'mean','duration_sector_3':'mean'})
albon_rain_avg = albon_rain_avg.reset_index()
albon_rain_avg = albon_rain_avg.merge(rain_race, on = 'session_key', how = 'inner')
albon_rain_avg
```

```python
leclerc_laps = pd.read_csv('../data/leclerc_laps.csv')
leclerc_laps = leclerc_laps[leclerc_laps.session_key.isin(rain_list)]
leclerc_rain = leclerc_laps[['session_key','driver_number','i1_speed','i2_speed','st_speed','lap_duration','duration_sector_1','duration_sector_2','duration_sector_3']]
leclerc_rain = leclerc_rain.reset_index(drop=True)
leclerc_rain_avg = leclerc_rain.groupby(['session_key']).agg({'i1_speed':'mean','i2_speed':'mean','st_speed':'mean','lap_duration':'mean','duration_sector_1':'mean','duration_sector_2':'mean','duration_sector_3':'mean'})
leclerc_rain_avg = leclerc_rain_avg.reset_index()
leclerc_rain_avg = leclerc_rain_avg.merge(rain_race, on = 'session_key', how = 'inner')
leclerc_rain_avg
```
