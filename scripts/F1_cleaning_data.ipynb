```python
import requests
import matplotlib.pyplot as plt
import json
import pandas as pd
```

```python
def url_to_df(url):
    endpoint = url
    response = requests.get(endpoint)
    response
    if response.status_code == 200:
        F1 = response.json()
        F1_df = pd.DataFrame(F1)
    else: print(response.status_code)
    return F1_df
```

```python
def agg_to_df(aggdf):
    x = aggdf.session_key.drop_duplicates().reset_index(drop = True)
    x = sessions.loc[sessions.session_key.isin(x)].reset_index(drop=True)
    return x
```

```python
sessions = pd.read_csv('../data/session_api.csv')
weather = pd.read_csv('../data/weather_api.csv')
drivers = pd.read_csv('../data/drivers_api.csv')
lando_laps = pd.read_csv('../data/laps_lando_api.csv')
albon_laps = pd.read_csv('../data/laps_albon_api.csv')
leclerc_laps = pd.read_csv('../data/laps_leclerc_api.csv')
```

```python
lando_laps = lando_laps.drop(columns = 'Unnamed: 0')
albon_laps  = albon_laps.drop(columns = 'Unnamed: 0')
leclerc_laps = leclerc_laps.drop(columns = 'Unnamed: 0')
```

```python
leclerc_laps
```

```python
sessions = sessions.drop(columns = 'Unnamed: 0')
```

```python
weather = weather.drop(columns = 'Unnamed: 0')
```

```python
drivers = drivers.drop(columns = ['Unnamed: 0'])
driver_session = drivers.merge(sessions, on = ['session_key'],how = 'inner')
```

```python
driver_session
```

```python
lap_list = [lando_laps,albon_laps,leclerc_laps]
all_laps = pd.concat(lap_list)
all_laps = all_laps.merge(driver_session, on = ['session_key','driver_number'], how = 'inner')
all_laps = all_laps.merge(weather, on = ['session_key','date_start'], how = 'inner')
```

```python
r = all_laps[all_laps.session_key == 9094]
r = r[r.driver_number == 4]
r
```

```python
track_temp_diff = all_laps.groupby(['session_key'])['track_temperature'].agg(lambda x: x.max() - x.min()).reset_index()
track_temp_diff.head()
```

```python
top_track_temp_diff = track_temp_diff.sort_values('track_temperature', ascending = False).head()
top_track_temp_diff 
```

```python
track_temp_list = top_track_temp_diff.session_key.to_list()
tt_diff = weather[weather.session_key.isin(track_temp_list)]
tt_diff
```

```python
lando_tt = lando_laps.merge(tt_diff, on = ['date_start','session_key'], how = 'inner')
lando_tt.head()
```

```python
albon_tt = albon_laps.merge(tt_diff, on = ['date_start','session_key'], how = 'inner')
albon_tt.head()
```

```python
leclerc_tt = leclerc_laps.merge(tt_diff, on = ['date_start','session_key'], how = 'inner')
leclerc_tt.head()
```

```python
ttframes = [leclerc_tt,albon_tt,lando_tt]
tt_comb = pd.concat(ttframes)
tt_comb
```

```python
weather = weather.groupby(['session_key']).agg({'air_temperature':'mean', 'humidity':'mean', 'pressure':'mean','rainfall':'max','track_temperature':'mean','wind_direction':'mean','wind_speed':'mean'})
```

```python
session_weather = sessions.merge(weather, on = ['session_key'], how = 'inner')
session_weather[['air_temperature','humidity','pressure','rainfall','track_temperature','wind_direction','wind_speed']].describe()
```

```python
rain_race = session_weather[session_weather.rainfall != 0]
rain_race
```

```python
hot_race = session_weather.sort_values('air_temperature', ascending = False).reset_index(drop=True).head(10)
hot_race
```

```python
cold_race = session_weather.sort_values('air_temperature', ascending = True).reset_index(drop=True).head(10)
cold_race
```

```python
windy_race = session_weather.sort_values('wind_speed', ascending = False).reset_index(drop=True).head(10)
windy_race
```

```python
avg_race = session_weather.iloc[(session_weather['air_temperature']-74.869).abs().argsort()[:15]]
```

```python
avg_race = avg_race[avg_race.rainfall == 0]
avg_race
```
