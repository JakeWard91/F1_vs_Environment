```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
weather = pd.read_csv('../data/weather.csv')
sessions= pd.read_csv('../data/sessions.csv')
weather = weather.merge(sessions, on = 'session_key', how = 'inner')
weather = weather[['session_key','location','year','air_temperature','humidity','pressure','rainfall','track_temperature','wind_direction','wind_speed']]
weather
```

```python
albon_laps = pd.read_csv('../data/albon_laps.csv') # reads in Albon's lap stats
albon_laps = albon_laps[['session_key','driver_number','i1_speed','i2_speed','st_speed','lap_duration','duration_sector_1','duration_sector_2','duration_sector_3']]
# reduces df to only necessary columns
albon_avg = albon_laps.groupby(['session_key','driver_number']).agg({'i1_speed':'mean','i2_speed':'mean','st_speed':'mean','lap_duration':'mean','duration_sector_1':'mean','duration_sector_2':'mean','duration_sector_3':'mean'})
# avgerages out all the lap data by each race
albon_avg = albon_avg.reset_index()
albon_avg
```

```python
leclerc_laps = pd.read_csv('../data/leclerc_laps.csv') # reads in Leclerc's lap stats
leclerc_laps = leclerc_laps[['session_key','driver_number','i1_speed','i2_speed','st_speed','lap_duration','duration_sector_1','duration_sector_2','duration_sector_3']]
# reduces df to only necessary columns
leclerc_avg = leclerc_laps.groupby(['session_key','driver_number']).agg({'i1_speed':'mean','i2_speed':'mean','st_speed':'mean','lap_duration':'mean','duration_sector_1':'mean','duration_sector_2':'mean','duration_sector_3':'mean'})
# avgerages out all the lap data by each race
leclerc_avg = leclerc_avg.reset_index()
leclerc_avg
```

```python
lando_laps = pd.read_csv('../data/lando_laps.csv') # reads in lando's lap stats
lando_laps = lando_laps[['session_key','driver_number','i1_speed','i2_speed','st_speed','lap_duration','duration_sector_1','duration_sector_2','duration_sector_3']]
# reduces df to only necessary columns
lando_avg = lando_laps.groupby(['session_key','driver_number']).agg({'i1_speed':'mean','i2_speed':'mean','st_speed':'mean','lap_duration':'mean','duration_sector_1':'mean','duration_sector_2':'mean','duration_sector_3':'mean'})
# avgerages out all the lap data by each race
lando_avg = lando_avg.reset_index()
lando_avg
```

```python
leclerc_avg.to_csv('../data/clean/leclerc_all.csv')
lando_avg.to_csv('../data/clean/lando_all.csv')
albon_avg.to_csv('../data/clean/albon_all.csv')
weather.to_csv('../data/clean/all_weather.csv')
```
