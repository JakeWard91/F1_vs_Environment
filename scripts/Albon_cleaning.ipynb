```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def type_to_df(df):
    df = df[['location','session_key','year','air_temperature','humidity','pressure','rainfall','track_temperature','wind_speed']]
    df_list = df.session_key.to_list()
    driver_df = albon_avg[albon_avg.session_key.isin(df_list)]
    driver_df = driver_df.merge(df, on = 'session_key', how = 'inner')
    return driver_df
```

```python
albon_laps = pd.read_csv('../data/albon_laps.csv') # reads in Albon's lap stats
albon_laps = albon_laps[['session_key','driver_number','i1_speed','i2_speed','st_speed','lap_duration','duration_sector_1','duration_sector_2','duration_sector_3']]
# reduces df to only necessary columns
albon_avg = albon_laps.groupby(['session_key','driver_number']).agg({'i1_speed':'mean','i2_speed':'mean','st_speed':'mean','lap_duration':'mean','duration_sector_1':'mean','duration_sector_2':'mean','duration_sector_3':'mean'})
# avgerages out all the lap data by each race
albon_avg = albon_avg.reset_index()
albon_avg
```

```python
rain_race = pd.read_csv('../data/rain_race.csv') # reads in csv of all the rain races
albon_rain = type_to_df(rain_race)
albon_rain
```

```python
hot_race = pd.read_csv('../data/hot_race.csv') # reads in csv of all the hot races
albon_hot = type_to_df(hot_race)
albon_hot
```

```python
cold_race = pd.read_csv('../data/cold_race.csv') # reads in csv of all the cold races
albon_cold = type_to_df(cold_race)
albon_cold
```

```python
windy_race = pd.read_csv('../data/windy_race.csv') # reads in csv of all the windy races
albon_windy = type_to_df(windy_race)
albon_windy
```

```python
avg_race = pd.read_csv('../data/avg_race.csv') # reads in csv of all the average weather races
albon_avg_weather = type_to_df(avg_race)
albon_avg_weather 
```

```python
albon_avg_weather.to_csv('../data/clean/albon_avg_weather.csv')
albon_windy.to_csv('../data/clean/albon_windy.csv')
albon_cold.to_csv('../data/clean/albon_cold.csv')
albon_hot.to_csv('../data/clean/albon_hot.csv')
albon_rain.to_csv('../data/clean/albon_rain.csv')
```
