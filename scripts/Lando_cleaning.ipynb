```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def type_to_df(df):
    df = df[['location','session_key','year','air_temperature','humidity','pressure','rainfall','track_temperature','wind_speed']]
    df_list = df.session_key.to_list()
    driver_df = lando_avg[lando_avg.session_key.isin(df_list)]
    driver_df = driver_df.merge(df, on = 'session_key', how = 'inner')
    return driver_df
```

```python
lando_laps = pd.read_csv('../data/lando_laps.csv') # reads in Lando's lap stats
```

```python
lando_laps = lando_laps[['session_key','driver_number','i1_speed','i2_speed','st_speed','lap_duration','duration_sector_1','duration_sector_2','duration_sector_3']]
# reduces df to only necessary columns
```

```python
lando_avg = lando_laps.groupby(['session_key','driver_number']).agg({'i1_speed':'mean','i2_speed':'mean','st_speed':'mean','lap_duration':'mean','duration_sector_1':'mean','duration_sector_2':'mean','duration_sector_3':'mean'})
# avgerages out all the lap data by each race
```

```python
lando_avg = lando_avg.reset_index()
lando_avg
```

```python
rain_race = pd.read_csv('../data/rain_race.csv') # reads in csv of all the rain races
lando_rain = type_to_df(rain_race)
lando_rain
```

```python
hot_race = pd.read_csv('../data/hot_race.csv') # reads in csv of all the hot races
lando_hot = type_to_df(hot_race)
lando_hot
```

```python
cold_race = pd.read_csv('../data/cold_race.csv') # reads in csv of all the cold races
lando_cold = type_to_df(cold_race)
lando_cold
```

```python
windy_race = pd.read_csv('../data/windy_race.csv') # reads in csv of all the windy races
lando_windy = type_to_df(windy_race)
lando_windy
```

```python
avg_race = pd.read_csv('../data/avg_race.csv') # reads in csv of all the average weather races
lando_avg_weather = type_to_df(avg_race)
lando_avg_weather 
```
