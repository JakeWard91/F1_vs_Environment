```python
import requests
import pandas as pd
```

# Combine Drivers & Sessions & Weather

```python
sessions = pd.read_csv('../data/APIs/session_api.csv')
sessions = sessions.drop(columns = 'Unnamed: 0')
drivers = pd.read_csv('../data/APIs/drivers_api.csv')
drivers = drivers.drop(columns = ['Unnamed: 0'])
weather = pd.read_csv('../data/APIs/weather_api.csv')
weather = weather.drop(columns = 'Unnamed: 0')
```

```python
driver_session = drivers.merge(sessions, on = ['session_key'],how = 'inner')
```

```python
dsw = driver_session.merge(weather, on = ['session_key'],how = 'inner')
dsw.head()
```

```python
dsw['date_record'] = pd.to_datetime(dsw.date_start)
dsw['date_record'] = dsw['date_record'].dt.strftime('%H:%M')
dsw = dsw.drop(columns = ['date_start'])
```

```python
dsw.to_csv('../data/Combined_CSVs/dsw.csv')
```

# Combine Lando/Albon/Leclerc Laps

```python
lando_laps = pd.read_csv('../data/APIs/laps_lando_api.csv')
lando_laps = lando_laps.drop(columns = 'Unnamed: 0')
albon_laps = pd.read_csv('../data/APIs/laps_albon_api.csv')
albon_laps  = albon_laps.drop(columns = 'Unnamed: 0')
leclerc_laps = pd.read_csv('../data/APIs/laps_leclerc_api.csv')
leclerc_laps = leclerc_laps.drop(columns = 'Unnamed: 0')
```

```python
lap_list = [lando_laps,albon_laps,leclerc_laps]
all_driver_laps = pd.concat(lap_list)
all_driver_laps
```

```python
all_driver_laps['date_record'] = pd.to_datetime(all_driver_laps.date_start)
all_driver_laps['date_record'] = all_driver_laps['date_record'].dt.strftime('%H:%M')
all_driver_laps.head()
```

```python
all_driver_laps.to_csv('../data/Combined_CSVs/driver_laps.csv')
```

---

# Combine All Driver Laps with DSW

```python
dsw_laps = all_driver_laps.merge(dsw, on = ['session_key','date_record','driver_number'], how = 'inner')
dsw_laps
```

```python
dsw_laps.to_csv('../data/Combined_CSVs/dsw_laps.csv')
```
