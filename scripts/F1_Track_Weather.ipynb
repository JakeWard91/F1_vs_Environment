```python
import requests
import pandas as pd
import numpy as np
```

```python
dsw = pd.read_csv('../data/Combined_CSVs/dsw_laps.csv') # reads in csv created in Combined CSVs Notebook
```

```python
r9094 = dsw[dsw.session_key == 9094] # limits df to only data from session 9094 (Monaco 2023 Race)
r9094.sort_values(['track_temperature']) # sorts the data by track temeratures
```

```python
race_weather = dsw.groupby(['session_key']).agg({'air_temperature':'mean', 'humidity':'mean', 'pressure':'mean','rainfall':'max','track_temperature':'mean','wind_direction':'mean','wind_speed':'mean'})
# returns the averages for each weather type parameter by session key(race)
```

```python
race_weather = race_weather.reset_index()
race_weather
```

```python
race_weather.describe() 
```

```python
rain = race_weather[race_weather.rainfall != 0] # returns df by session key (race) where rain equals 1 (indicates it was raining)
rain
```

```python
hot = race_weather.sort_values('air_temperature', ascending = False).reset_index(drop=True).head(10)
# returns df's top 10 hottest air temperature by session key (race)
hot
```

```python
cold = race_weather.sort_values('air_temperature', ascending = True).reset_index(drop=True).head(10)
# returns df's top 10 coldest air temperature by session key (race)
cold
```

```python
wind = race_weather.sort_values('wind_speed', ascending = False).reset_index(drop=True).head(10)
# returns df's top 10 windiest by session key (race)
wind
```

```python
avg_race = race_weather.iloc[(race_weather['air_temperature']-74.869).abs().argsort()[:15]]
# returns df's most average races by air temp by finding the mean and using it to find races with deviation of less than 15 degrees
avg_race = avg_race[avg_race.rainfall == 0] # returns previous df where it was also not raining
avg_race
```

```python
track_temp_diff = dsw.groupby(['session_key'])['track_temperature'].agg(lambda x: x.max() - x.min()).reset_index()
# returns the session keys (races) and the difference in the min and max track temperatures 
track_temp_diff.head()
```

```python
top_track_temp_diff = track_temp_diff.sort_values('track_temperature', ascending = False).head()
# returns the previous df's top 5 races with the largest track temp variance
top_track_temp_diff 
```
